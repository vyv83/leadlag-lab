# leadlag exploration

Use this notebook to inspect analyzed sessions, filter events, and decide which follower/hold/spread combinations deserve a strategy.

```python
from leadlag import list_sessions, load_session

sessions = list_sessions('../data')
sessions[:5]
```

```python
session_id = sessions[0]['id']
s = load_session(session_id, data_dir='../data')
s.meta
```

```python
events = s.events
events.count, events.filter(signal='C').count
```

```python
follower = s.meta['followers'][0]
events.filter(signal='C', follower=follower).stats(follower)
```

```python
events.plot_magnitude_distribution()
```

```python
events.plot_lag_distribution(follower)
```

```python
events.filter(signal='C', follower=follower)[0].plot(follower=follower)
```

```python
events.filter(signal='C').plot_heatmap(
    x='delay_ms', y='hold_ms', metric='net_pnl_bps', follower=follower
)
```
